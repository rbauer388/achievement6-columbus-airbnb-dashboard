```python
# Import Json

import json

geo_path = r"C:\Users\rbaue\Desktop\A6_AirBnB_Cleaning\us_states.geojson.json"

with open(geo_path, "r", encoding="utf-8") as f:
    geo_data = json.load(f)

geo_data.keys()
```

```python
# Wrangling

df["price"] = (
    df["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
# Aggregate to neighborhood

neigh = (
    df.groupby("neighbourhood", as_index=False)
      .agg(avg_price=("price", "mean"),
           count_listings=("id", "count"))
)

neigh.head()
```

```python
# Filter out some neighborhoods

neigh = neigh[neigh["count_listings"] >= 20]
```
